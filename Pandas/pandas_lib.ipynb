# **Series**
Series is a pandas 1-Dimensional labeled array that can hold any data type. It likes a single column in a spreadsheet (1-Dimensional)

```python
data = ['A', 'B', 'C', 2]
series = pd.Series(data)
series.loc[1] = 'F'
print(series.loc[0])
print(series)
```

```python
label = [chr(i) for i in range(ord('a'), ord('e')+1)]
print(label)
```

**Example: Calories**
1. Change value of a location
2. Filter by value

```python
calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}
calories_series = pd.Series(calories)

calories_series.loc["Day 3"] += 500
print(calories_series.loc["Day 3"])
print(calories_series)

print(f'In which day he did not stick to his diet: {calories_series[calories_series > 2000]}')
```

# **DataFrame**

```python
dataframe = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age": [30, 35, 50]
}

df = pd.DataFrame(dataframe, index=['Employee 1', 'Employee 2', 'Employee 3'])
print(df)
print('-------')
print(df.loc['Employee 1'])
print('-------')
print(df.iloc[0])
```

* Add a new column: create a python list, the number of value must match the number of row
* Add a new row: create a new DataFrame then append in the previous

```python
df["Job"] = ["Cook", "N/A", "Cashier"]
print(df)
```

```python
new_row = pd.DataFrame([{"Name": "Sandy", "Age": 28, "Job": "Engineer"},
                        {"Name": "Eugene", "Age": 60, "Job": "Manager"}], 
                       index=["Employee 4", "Employee 5"])
df = pd.concat([df, new_row])
print(df)
```

# **Filtering**
Keep the rows that match a condition

```python
tall_pokemon = df3[df3["Height"] >= 2]
print(f"Tall Pokemon: {tall_pokemon}")
print("--------------------------------------------")
```

```python
heavy_pokemon = df3[df3["Weight"] > 100]
print(f"Heavy Pokemon: {heavy_pokemon}")
print("--------------------------------------------")
```

```python
legendary = df3[df3["Legendary"] == True]
print(f"legendary Pokemon: {legendary}")
```

# **Aggregation**
Reduces a set of values into a single summary value  
Used to summarize and analyze data  
Often used with groupby() function


**Whole DataFrame**

```python
print("------------Sum-------------")
print(df3.sum(numeric_only=True))
print("------------Mean------------")
print(df3.mean(numeric_only=True))
print("------------Min-------------")
print(df3.min(numeric_only=True))
print("------------Max-------------")
print(df3.max(numeric_only=True))
```

**Single Column**

```python
print(df3["Type1"].unique())
print(df3["Type1"].count())
```

```python
group1 = df2.groupby("Type1")
group1_mean = group1["Height"].mean()
df_group1_mean = group1_mean.reset_index()
print(type(df_group1_mean))

print(group1["Height"].count())
```
