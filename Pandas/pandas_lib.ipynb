# **Series**
Series is a pandas 1-Dimensional labeled array that can hold any data type. It likes a single column in a spreadsheet (1-Dimensional)

```python
data = ['A', 'B', 'C', 2]
series = pd.Series(data)
series.loc[1] = 'F'
print(series.loc[0])
print(series)
```

```python
label = [chr(i) for i in range(ord('a'), ord('e')+1)]
print(label)
```

**Example: Calories**
1. Change value of a location
2. Filter by value

```python
calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}
calories_series = pd.Series(calories)

calories_series.loc["Day 3"] += 500
print(calories_series.loc["Day 3"])
print(calories_series)

print(f'In which day he did not stick to his diet: {calories_series[calories_series > 2000]}')
```
