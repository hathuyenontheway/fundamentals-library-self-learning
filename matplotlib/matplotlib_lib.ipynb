```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# **Plot Customization**


```python
x = np.array([2023,2024,2025,2026])
y1 = np.array([15,25,30,20])
y2 = np.array([8,5,40,12])
y3 = np.array([30,50,45,23])

line_style = dict(marker=".", markersize=10, markerfacecolor="cyan", linestyle="solid", linewidth=2,)
```

# **Labels**

```python
plt.title("Class Size", fontsize=15, fontweight="bold", color="blue")
plt.xlabel("Year", fontsize=15, family="Arial", fontweight="bold", color="#2dbefc")
plt.ylabel("Student", fontsize=15, family="Arial", fontweight="bold", color="#2dbefc")

plt.figure(1)
plt.plot(x, y1, **line_style)
plt.plot(x, y2, color="green", **line_style)
plt.plot(x, y3, color="red",**line_style)

plt.tick_params(axis='both', colors='red')

plt.xticks(x)
plt.show()
```

```python
plt.figure(2)
plt.plot(x, y1)
plt.grid(axis='x')

plt.show()
```

# **Bar Chart**
Compare categories of data by representing each category with a bar

```python
categories = np.array(['Grains', 'Fruit', 'Vegetables', 'Protein', 'Dairy', 'Sweets'])
values = np.array([4,3,1,5,3,2])

plt.title("Daily Consumption", color="black")
plt.xlabel("Food", color="darkblue")
plt.ylabel("Quanlity", color='darkblue')

plt.bar(categories, values, color='skyblue')
plt.show()
```

# **Pie Chart**
Circular chart divided into slices to show percentages of the total. Good for visualizing distribution among categories

```python
pie_categories = np.array(["Freshmen", "Sophomores", "Juniors", "Seniors"])
values = np.array([300,250,275,225])

colors = ["red", "yellow", "blue", "green"]

plt.pie(values, labels=pie_categories, autopct="%1.1f%%", colors=colors, explode=[0.1,0.1,0.1,0.1], shadow=True)
plt.title("Bro Code Collage", fontweight="bold")

plt.show()
```

# **Scatter Graph**
Shows the relationship between two variables  
Helps to identify a correlation (+, -, None)  
Example: Study hours vs. Test scores  

```python
sca_x = np.array([0, 1, 1, 2, 3, 4, 6, 7, 7, 8]) # Hours studied
sca_y = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 87]) # Grades

plt.scatter(sca_x, sca_y, color="darkblue", alpha=0.5)
plt.xlabel("Hours Studied")
plt.ylabel("Grade")
plt.title("Test Scores", fontweight="bold")
plt.show()
```
